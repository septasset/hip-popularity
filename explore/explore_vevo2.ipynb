```python
### examine the view pattern
days_views = []
for i, line in enumerate(vevo_json):
    vevo_record = json.loads(line)
    days = [int (x) for x in vevo_record['insights']['days'].split(",")]
    views = [int(x) for x in vevo_record['insights']['dailyView'].split(",")]
    if len(days) >= 489 and len(days) <= 1175:
        days_views.append((days, views))
```

```python
for i in range(0, 8):
    fig = plt.figure(figsize=(24, 6))
    ax = fig.add_subplot(111)
    days, views = days_views[i]
    ax.scatter(days, views)
```
